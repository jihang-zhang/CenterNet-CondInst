```python
import json
import pandas as pd
```

```python
df = pd.read_csv('../../input/train.csv')
```

```python
# replace nan values with zeros
df['bbox'] = df.bbox.fillna('[0,0,0,0]')

# split bbox column
bbox_items = df.bbox.str.split(',', expand=True)
df['bbox_xmin'] = bbox_items[0].str.strip('[ ').astype(float)
df['bbox_ymin'] = bbox_items[1].str.strip(' ').astype(float)
df['bbox_width'] = bbox_items[2].str.strip(' ').astype(float)
df['bbox_height'] = bbox_items[3].str.strip(' ]').astype(float)
```

```python
df['aspect_ratio'] = df['bbox_width'] / df['bbox_height']
```

```python
df['area'] = df['bbox_height'] * df['bbox_width']
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df['wh'] = (df['bbox_width'] / df['bbox_height']).round()
```
